The data was downloaded using OSMnx, here we work with the already downloaded data, to work with new one please check [OSMnx documentation](https://osmnx.readthedocs.io/en/stable/)

This notebook uses the data processed by notebook 02_BicycleAlgorithms.ipynb and reproduces the results in the fifth section of the paper: *Natera, L. et al "Devoloping optimal bicycle network growth strategies"*

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import count
import pandas as pd
import matplotlib as mpl
import osmnx as ox
import matplotlib.cm as cm
import matplotlib.colors as mpcol
%matplotlib inline
```

```python
algorithms = ['L2C','L2S','CC','R2C']
```

```python
name = 'Budapest'

title_size = 15
label_size = 33
major_size = 15
minor_size = 15
figure_title = 40
max_len = 0
xmin=-3
xmax=80


shapes = ['.', '*', '+', 'x'] 
colors=['#ff5a5f','#0a2463','#25283d', '#333333']#'#087e8b',
styles = ['-','--','-.',':']
widths = [3.5,2.5,1.5,1.5]
```

```python
handles=[]
fig, axes = plt.subplots(2,2,figsize=(20,16),constrained_layout=True, sharex=True)
for al, shape, color, width,style in zip(algorithms, shapes, colors, widths, styles):
    df = pd.read_csv('outputs/CSV/{}_{}.csv'.format(name, al), index_col=0)
    df['directness'] = df['d_ij_b'] / df['d_ij_s']
    df.loc[0:2, 'Gain_km'] = 0
    for i in range(1, len(df)):
        df.loc[i, 'Gain_km'] = ((df.loc[i,'length_cc'] - df.loc[1, 'length_cc']) / df.loc[1, 'length_cc'])
    
    handles.append(axes[0,0].plot(df['delta']/1000, df['nodes_cc']/df['nodes_cc'].max(), color=color, label=al, alpha=0.75, linewidth=width,linestyle=style)[0])
    axes[0,1].plot(df['delta']/1000, df['length_cc']/df['length_cc'].max(), color=color, label=al,  alpha=0.75, linewidth=width,linestyle=style)
    axes[1,0].plot(df['delta']/1000, df['Gain_km'], color=color, label=al, alpha=0.75, linewidth=width,linestyle=style)
    axes[1,1].plot(df['delta']/1000, df['directness'], color=color, label=al, alpha=0.75, linewidth=width,linestyle=style)

x_label = fig.text(0.5, -0.02, r'$\Delta_x\ (km)$', va='center', ha='center', fontsize=40, transform=fig.transFigure)
axes[0,0].set_ylabel('Fraction of nodes\n$n_{LCC}$', fontsize=label_size)
axes[0,1].set_ylabel('Fraction of link\nkilometers $\ell_{LCC}$', fontsize=label_size)
axes[1,0].set_ylabel('Kilometers gain $G_{LCC}$', fontsize=label_size)
axes[1,1].set_ylabel('Bike-car\ndirectness $\Delta$', fontsize=label_size)

for ax in axes.flat:
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    #ax.spines['left'].set_smart_bounds(True)
    ax.spines['bottom'].set_smart_bounds(True)
    ax.tick_params(labelsize=20)
    
axes[1,1].set_ylim(0,1)
    
fig.legend(handles[0:4],algorithms, loc=4, bbox_to_anchor=[1.0, 0.08], fontsize=20, markerscale=6, frameon=False, title='Algorithms', title_fontsize=25, ncol=1)

plt.savefig('outputs/imgs/AlgorithmsMeasures.png',bbox_include=[x_label],constrained_layout=True,bbox_inches = 'tight', pad_inches = 0.5);
```
